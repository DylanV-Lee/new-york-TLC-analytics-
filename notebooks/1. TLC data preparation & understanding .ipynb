**Lets look into the payment_type column**

Note: In the dataset, payment_type is encoded in integers:

- 1: Credit card
- 2: Cash
- 3: No charge
- 4: Dispute
- 5: Unknown

```python
# What is the total count of each payment type in payment_type variable

data['payment_type'].value_counts()
```

```python
# what is the average tip for trips paid for with credit card

avg_tip_by_creditcard = data[data['payment_type']==1]['tip_amount'].mean()
print(f'The average tip for amount paid with credit card: {avg_tip_by_creditcard}')
```

```python
# what is the average tip paid with cash

avg_tip_with_cash = data[data['payment_type']==2]['tip_amount'].mean()
print(f'The average tip for amount paid with credit card: {avg_tip_with_cash}')
```

```python
# how many different vendors are there?

data['VendorID'].value_counts()
```

```python
# what is the mean of total amount for each vendor?

data.groupby(['VendorID']).mean(numeric_only = True)['total_amount']
```

```python
# we might need to analyse the data with different payment type

# filter data for credit card
credit_card = data[data['payment_type'] == 1]
```

```python
# lets look in the passenger count for payments paid with credit cards

credit_card['passenger_count'].value_counts()
```

```python
# what is the average tip for each passenger count

credit_card.groupby(['passenger_count']).mean(numeric_only = True)['tip_amount']
```
